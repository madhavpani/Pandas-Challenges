```python
# import the pandas library with alias pd to work with it
import pandas as pd
```

1. Write a Pandas program to display the dimensions or shape of the World alcohol consumption dataset. Also extract the column names from the dataset.


```python
# i have downloaded the csv file 'world_alcohol.csv' 
# read the csv file using pd.read_csv(file_path)
df = pd.read_csv('world_alcohol.csv')
print('first 5 rows:')
print(df.head())
print('\nshape: ')
print(df.shape)
print('\ncolumns: ')
print(df.columns)
```

2. Write a Pandas program to select first 2 rows, 2 columns and specific two columns from World alcohol consumption dataset.


```python
df = pd.read_csv('world_alcohol.csv')

# df.iloc[rows, cols] : This property allows us to access the specified rows and cols from the df.
df.iloc[:2, :2]
```

3. Write a Pandas program to select random number of rows, fraction of random rows from World alcohol consumption dataset.


```python
df = pd.read_csv('world_alcohol.csv')


# df.iloc[rows, cols] : This property allows us to access the specified rows and cols from the df.
# df.sample(n) : This will select the random n rows from the df.
# df.sample(frac=n) : This will select the n percentage of random rows out of all rows.

print('Random 4 rows: ')
print(df.sample(4))
print('\nRandom 10% rows: ')
print(df.sample(frac=.1))
```

4. Write a Pandas program to find and drop the missing values from World alcohol consumption dataset.


```python
df = pd.read_csv('world_alcohol.csv')

print('null values: ')
print(df.isna())

# df.dropna() : This method drops all the null values from the df.
print('Dropped all missing values: ')
print(df.dropna())
```

5. Write a Pandas program to remove the duplicates from 'WHO region' column of World alcohol consumption dataset.


```python
df = pd.read_csv('world_alcohol.csv')

# df[col].drop_duplicates() : This method drops all the duplicates from the specified column from the df.
df['WHO region'].drop_duplicates()
```

7. Write a Pandas program to find out the alcohol consumption details by the 'Americas' in the year '1985' from the world alcohol consumption dataset.
