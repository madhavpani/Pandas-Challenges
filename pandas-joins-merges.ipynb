1. Write a Pandas program to join the two given dataframes along rows and assign all data.

4. Write a Pandas program to join the two given dataframes along rows and merge with another dataframe along the common column id.

```python
# lets create two dfs: 
df1 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
         'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'], 
        'marks': [200, 210, 190, 222, 199]})

df2 = pd.DataFrame({
        'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
        'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'], 
        'marks': [201, 200, 198, 219, 201]})

df3 = pd.DataFrame({
        'student_id': ['S1', 'S2', 'S3', 'S4', 'S5', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'S13'],
        'exam_id': [23, 45, 12, 67, 21, 55, 33, 14, 56, 83, 88, 12]})

# pd.concate(df_list, axis, ignore_index) : this method concates the dfs specified along the given axis.
# axis=0 means along rows
# axis=1 means along cols
df4 = pd.concat([df1, df2], axis=0, ignore_index=True)

# pd.merge(left, right, on, how, suffixes) : This is a method used to merge two df or series into one.
# parameters: 
# left : it is a df, which is on left side
# right : it is a df, which is on right side
# how : it has valus like [ inner, outer, left, right, cross ], which specifies the type of merging
# on : this takes the common column to merge both the df
# suffixes : This is the suffix to add the column_name if more than 2 common column_name is found.

pd.merge(left=df4, right=df3, on='student_id')
```

7. Write a Pandas program to join (left join) the two dataframes using keys from left dataframe only.


```python
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 

data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})

pd.merge(data1, data2, how='left', on=['key1', 'key2'])
```

11. Write a Pandas program to combine the columns of two potentially differently-indexed DataFrames into a single result DataFrame.

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                      'B': ['B0', 'B1', 'B2']},
                     index=['K0', 'K1', 'K2'])

df2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])

# df1.join(df2) : This method joins two dfs.
df1.join(df2)
```
