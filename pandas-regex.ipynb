1. Write a Pandas program to convert all the string values to upper, lower cases in a given pandas series. Also find the length of the string values.


```python
# lets create a series
series = pd.Series(['X', 'Y', 'Z', 'Aaba', 'Baca', np.nan, 'CABA', None, 'bird', 'horse', 'dog'])
print('Original Series: \n', series)

# series.str : This is the core module to work with strings
# series.str.upper() : This method converts the values of the series to uppercase.
print('Upper Case: ')
print(series.str.upper())

# series.str.lower() : this method converts the values of the series to lowercase.
print('Lower Case: ')
print(series.str.lower())
```

12. Write a Pandas program to extract email from a specified column of string type of a given DataFrame.


```python
# lets import regular expression library of python
import re

df = pd.DataFrame({
    'name_email': ['Alberto Franco af@gmail.com','Gino Mcneill gm@yahoo.com','Ryan Parkes rp@abc.io', 'Eesha Hinton', 'Gino Mcneill gm@github.com']
    })

# the pattern variable contains the pattern of email using regular expression
pattern = r'[\w\.-]+@[\w\.-]+'

# df[col].apply() : This method applies the lambda function to all the values of the column.
df['email'] = df['name_email'].apply(lambda x: re.findall(pattern, x))
df['email'] = df['email'].apply(lambda x: x[0] if x else '')
df
```

21. Write a Pandas program to extract numbers greater than 940 from the specified column of a given DataFrame.


```python
df = pd.DataFrame({
    'company_code': ['c0001','c0002','c0003', 'c0003', 'c0004'],
    'address': ['7277 Surrey Ave.1111','920 N. Bishop Ave.','9910 Golden Star St.', '1025 Dunbar St.', '1700 West Livingston Court']
    })

pattern = r'9[456789]\d+|\d{4}'

df['gt940'] = df['address'].apply(lambda x: re.findall(pattern, x))
df['gt940'] = df['gt940'].apply(lambda x: ' '.join(x) if x else '')
df
```

22. Write a Pandas program to extract date (format: mm-dd-yyyy) from a given column of a given DataFrame.


```python
df = pd.DataFrame({
    'company_code': ['Abcd','EFGF', 'zefsalf', 'sdfslew', 'zekfsdf'],
    'date_of_sale': ['12/05/2002','16/02/1999','05/09/1998','12/02/2022','15/09/1997'],
    'sale_amount': [12348.5, 233331.2, 22.5, 2566552.0, 23.0]
})

# date pattern
pattern = r'([0-2]\d|3[0-1])\/(0[1-9]|1[1-2])\/(\d{4})'

df['date'] = df['date_of_sale'].apply(lambda x: re.findall(pattern, x))
df['date'] = df['date'].apply(lambda x: x[0] if x else '')
df
```
